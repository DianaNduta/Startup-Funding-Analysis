## Dealing with outliers

> Dealing with the extreme or unusual data points that can adversely impact the model's performance.

Based on the empirical rule or the 68-95-99.7 rule. According to this rule, in a normal distribution, approximately 99.7% of the data falls within 3 standard deviations from the mean. By using a factor of 3, we are considering values that are relatively far from the mean as potential outliers.

```python
# Convert DataFrame to numeric type
data_numeric = data.apply(pd.to_numeric, errors='coerce')

# Calculate the upper and lower thresholds for outlier detection
upper_threshold = 3 * data_numeric.std() + data_numeric.mean()
lower_threshold = data_numeric.mean() - 3 * data_numeric.std()


# Replace the outliers with the column's median
data_numeric = data_numeric.apply(lambda x: x.where(~((x > upper_threshold[x.name]) | (x < lower_threshold[x.name]))), axis=0)

# Check if any outliers are still present
outliers_present = (data_numeric > upper_threshold) | (data_numeric < lower_threshold)

# Return True if any outliers are found, False otherwise
data_contains_outliers = outliers_present.any().any()
print(data_contains_outliers)
```

### creating a function that takes in our dataframe and returns corellations between column in pair in descending order

```python
def get_correlation_df(df):
    corr_df = df.corr().abs().stack().reset_index().sort_values(0, ascending=False)
    corr_df['pairs'] = list(zip(corr_df.level_0, corr_df.level_1))
    corr_df.set_index(['pairs'], inplace=True)
    corr_df.drop(columns=['level_1', 'level_0'], inplace=True)
    corr_df.columns = ['cc']
    corr_df = corr_df.drop_duplicates()
    return corr_df.head(50)

get_correlation_df(df2)
```

## creating a function to print_evaluation_results

```python
def print_evaluation_results(Y_test, y_pred):
    print("MODEL RESULTS")
    print('Accuracy: ', round(accuracy_score(Y_test, y_pred), 5))
    print('F1 score: ', round(f1_score(Y_test, y_pred, average='macro'), 5))
    print('Recall: ', round(recall_score(Y_test, y_pred, average='macro'), 5))
    print('Precision: ', round(precision_score(Y_test, y_pred, average='macro'), 5))
    cm_dt = confusion_matrix(Y_test, y_pred)
    f, ax = plt.subplots(1, 1, figsize=(5, 3))
    sns.heatmap(cm_dt, annot=True, cmap='Greens', fmt='g', ax=ax)
    ax.set_xlabel('Predicted Labels')
    ax.set_ylabel('True Labels')
    ax.set_title('Confusion Matrix')
    ax.xaxis.set_ticklabels(['0', '1', '2'])
    ax.yaxis.set_ticklabels(['0', '1', '2'])
    plt.show()
```

# MODEL 4 - AdaBoosting

AdaBoost, short for Adaptive Boosting, is a popular ensemble learning algorithm that combines multiple weak classifiers to create a strong classifier. It is particularly effective in handling binary classification problems, but it can also be extended to handle multi-class classification tasks. AdaBoost works by iteratively training weak classifiers on different subsets of the training data, focusing more on the misclassified samples in each iteration.

In each iteration, AdaBoost assigns weights to each training sample based on their classification performance. Initially, all samples are assigned equal weights, and the algorithm seeks to find a weak classifier that performs better than random guessing. After each iteration, the weights of the misclassified samples are increased, forcing subsequent weak classifiers to pay more attention to those samples. This adaptive weighting scheme allows AdaBoost to prioritize the difficult instances and focus on improving the overall classification accuracy.

The final classification model in AdaBoost is a weighted combination of all the weak classifiers, where each classifier's weight is determined by its performance in the training process. During the prediction phase, each weak classifier's output is weighted based on its performance, and the weighted votes are combined to obtain the final prediction. This ensemble approach enhances the overall accuracy and robustness of the model, making AdaBoost particularly useful in scenarios where individual classifiers may struggle.

# Applying SFS (Sequential Feature Selector) Feature Selection Technique to our Random Forest model

Sequential Feature Selector (SFS) is a feature selection technique that iteratively selects the most relevant features for a given task. It reduces dimensionality and improves model performance by choosing features based on predefined criteria. SFS explores different feature combinations and evaluates their impact on model performance. It improves interpretability and computational efficiency by selecting informative features.

During each iteration, SFS evaluates different subsets of features by training and testing a machine learning model. It considers both the individual performance of features and their interactions with other selected features. This way, SFS explores different combinations of features to identify the most informative subset.
